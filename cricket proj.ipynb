### displaying the graph for matches per year

```python
df['Year'] = df['Match Date'].str[-4:]
yearwise = df[['Year']].value_counts()/2
yearwise = yearwise.reset_index()
yearwise.columns =['Year', 'Matches']
yearwise =yearwise.sort_values(by='Year', ascending=True)
sns.set_style('darkgrid')
plt.figure(figsize=(20,10))
plt.bar(yearwise['Year'], yearwise['Matches'], width=0.65, color='Green')
plt.xticks(rotation='vertical', size=15)
plt.yticks(size=15)
plt.xlabel('Year', size=20)
plt.ylabel('No. of Matches', size=20)
plt.title('Matches per year', size=25)
plt.show
```

### concatenation of team1 and team2  into single teams

```python
group1=df.groupby("Team 1").size().rename("Team 1")
group2=df.groupby("Team 2").size().rename("Team 2")
total=pd.concat([group1,group2],axis=1).fillna(0)
total["total"]=total["Team 1"]+total["Team 2"]
total.columns
```

### displaying the graph of odi matches played by team 

```python
sns.barplot(x=total.index,y=total["total"])
plt.ylabel("Number of Matches")
plt.xlabel("Teams")
plt.yticks([100,200,300,400,500,600,700,800])
plt.xticks(rotation=90)
plt.title("ODI Matches Played by team")
```

### converting categorical values into encoded values

```python
dataset_filtered =df[ (df['Team 1'].isin(['Australia','India'])) & (df['Team 2'].isin(['Australia','India']))]
x= dataset_filtered.drop(['Winner','Scorecard','Match Date'],axis =1)
y= dataset_filtered['Winner']
x_filtered = pd.get_dummies(x,drop_first = True)
y_filtered = pd.get_dummies(y,drop_first = True)
```

### split the data into training,test and validation sets 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_filtered,y_filtered,test_size=0.25)
```
